Question 14: Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temerature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data creation
date_range = pd.date_range(start='2024-01-01', end='2024-01-10', freq='D')
temperature = np.random.uniform(low=20, high=30, size=len(date_range))
humidity = np.random.uniform(low=30, high=90, size=len(date_range))

# Create a DataFrame
time_series_df = pd.DataFrame({
    'date': date_range,
    'temperature': temperature,
    'humidity': humidity
})

# Create the figure and first axis for temperature
fig, ax1 = plt.subplots()

# Plot temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(time_series_df['date'], time_series_df['temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(time_series_df['date'], time_series_df['humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Show the plot
fig.tight_layout()
plt.xlabel("Date")
plt.title('Temperature and Humidity Over Time')
plt.show()
```

Question 15: Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution (mean=0, std=1)
data = np.random.randn(1000)

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Create a range of x values
x = np.linspace(min(data), max(data), 1000)

# Compute the PDF of the normal distribution
pdf = norm.pdf(x, loc=0, scale=1)

# Plot the PDF as a line plot
plt.plot(x, pdf, 'r', label='Normal PDF')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Question 17: Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)  # Set a seed for reproducibility
x = np.random.randn(1000)
y = np.random.randn(1000)

# Determine quadrant for each data point
quadrants = np.zeros(len(x))
quadrants[(x >= 0) & (y >= 0)] = 1
quadrants[(x < 0) & (y >= 0)] = 2
quadrants[(x < 0) & (y < 0)] = 3
quadrants[(x >= 0) & (y < 0)] = 4

# Create scatterplot with color-coded quadrants
plt.figure(figsize=(8, 8))
plt.scatter(x, y, c=quadrants, cmap='viridis', alpha=0.7)

# Add quadrants
plt.axvline(0, color='black', linestyle='--')
plt.axhline(0, color='black', linestyle='--')

# Add legend
plt.legend(['Quadrant 1', 'Quadrant 2', 'Quadrant 3', 'Quadrant 4'])

# Label axes and set title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant Wise Scatterplot')

# Equal aspect ratio for a square plot
plt.axis('equal')

plt.show()
```

Question 18:With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'





```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate x values
x = np.linspace(0, 2 * np.pi, 100)

# Compute y values (sine of x)
y = np.sin(x)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Sine Wave Function", x_axis_label='x values', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the results
show(p)

```

Question 19: Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)  # Random integer values

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Set up the data source for Bokeh
source = ColumnDataSource(data)

# Define color mapping for bars based on the values
colors = ['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1', '#fddbc7']
mapper = factor_cmap(field_name='categories', palette=colors, factors=categories)

# Create a figure for the bar chart
p = figure(x_range=categories, height=400,width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Create bar chart
p.vbar(x='categories', top='values', width=0.8, source=source, legend_field="categories",
       line_color='white', fill_color=mapper)

# Add tooltips to display values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Customize the chart appearance
p.title.align = "center"
p.title.text_font_size = "20px"
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
output_file("random_categorical_bar_chart.html")
show(p)
```

Question 21:Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'
